```python
import numpy as np
import random
deck=[]                      
for i in range(4):                 #create 1 to 13 four times
    for i in range(1,14):
        deck.append(i)
# Define function to deal cards to two players
def deal():
    # Initialize player hands
    p1 = []
    p2 = []
    
    # Initialize and shuffle deck
    deck = list(range(1, 14)) * 4
    np.random.shuffle(deck)
    
    # Deal cards to each player
    for i in range(len(deck)):
        if i % 2:
            p2.append(deck.pop(0))
        else:
            p1.append(deck.pop(0))
            
    return p1, p2
def play(games=1): 
    initial_total1=[]
    initial_total2=[]
    final_total1=[]
    final_total2=[]
    num_ties=[]
    largest_total_won=[]
    for m in range(games):
        hand1,hand2=deal()
        score1=0
        score2=0
        i=1
        player1_round=[]                                
        player2_round=[]
        player1_cards=[]
        player2_cards=[]
        ties=0
        total_won=[]
        while i <=13:                                   #totally 13 rounds
            player1_round.append(hand1[i*2-2])          #player#_round used to record scores for each round
            player1_round.append(hand1[i*2-1])
            player2_round.append(hand2[i*2-2])
            player2_round.append(hand2[i*2-1])
            player1_cards.append(hand1[i*2-2])          #player#_cards used to record cards for each round
            player1_cards.append(hand1[i*2-1])
            player2_cards.append(hand2[i*2-2])
            player2_cards.append(hand2[i*2-1])
            sum1=sum(player1_round)                     #calculate the sum of the 2 cards of one roung for each player                  
            sum2=sum(player2_round)
            score=sum1+sum2                             #calculate the score for this round
            total_won.append(score)
            if sum1>sum2:                               #if player1 wins this round, add the score to score1
                score1+=score
                i+=1
                player1_round=[]                        #clean the list to record the next round
                player2_round=[]

            elif sum1<sum2:                             #same as the condition of sum1>sum2 
                score2+=score
                i+=1
                player1_round=[]
                player2_round=[]
            else:                                       #if ties for this round, just go to the next round and doesn't clean
                ties+=1
                if i <13:                               #the list so that next round still contain cards for this round
                    i+=1
                else:                                   #if the final round ties, return cards to each player
                    i+=1
                    score1+=sum1
                    score2+=sum2                  
        if score1>score2:                               #define the final winner
            winner='player1'
        elif score1<score2:
            winner='player2' 
        else:
            winner='Tie'
        initial_total1.append(sum(hand1))
        initial_total2.append(sum(hand2))
        largest_total_won.append(max(total_won))
        num_ties.append(ties)
        final_total1.append(score1)
        final_total2.append(score2)
    S={'Initial_value_totals_for_p1':initial_total1,'Initial_value_totals_for_p2':initial_total2,'Number_of_ties':num_ties,'Largest_total_won':largest_total_won,'Final_value_totals_for_p1':final_total1,'Final_value_totals_for_p2':final_total2}
        
    return S
```

c.Given your DataFrame of simulated game results, let's determine the outcome of each game, based on the final player totals. Add a new column to your DataFrame ('outcome') that contains the value 'Player 1 wins' if Player 1's final total is higher, 'Player 2 wins' if Player 2's final total is higher, or 'Tie game' if the two player's final totals are equal. In addition, add new columns for the winning total ('winning_total') and losing total ('losing_total'), which contain the final totals for the winning and losing players, respectively, and missing values (NumPy.nan) for tie games. (2 points)

Preview the first 10 rows of your modified DataFrame. Then, summarize the frequency of the three outcomes and display the rows in your DataFrame that contain the 'Tie game' outcomes. Finally, produce descriptive statistics (e.g., minimum, mean, standard deviation, maximum) for your winning total column, and display the rows in your DataFrame that contain the top 10 winning totals. (1 point)

```python
#Add new columns
df.loc[df['Final_value_totals_for_p1'] > df['Final_value_totals_for_p2'], 'outcome'] = 'Player 1 wins'
df.loc[df['Final_value_totals_for_p1'] < df['Final_value_totals_for_p2'], 'outcome'] = 'Player 2 wins'
df.loc[df['Final_value_totals_for_p1'] == df['Final_value_totals_for_p2'], 'outcome'] = 'Tie game'
import numpy as np
def functionwin(a, b):
    if a == b:
        return np.nan
    else:
        return max(a,b)
def functionlose(a, b):
    if a == b:
        return np.nan
    else:
        return min(a,b)
df['winning_total'] = df.apply(lambda x:functionwin(x.Final_value_totals_for_p1,x.Final_value_totals_for_p2), axis = 1)
df['losing_total'] = df.apply(lambda x:functionlose(x.Final_value_totals_for_p1,x.Final_value_totals_for_p2), axis = 1)
```

Then, calculate the mean maximum round total as a function of the number of ties. What do you observe about the relationship between the number of ties and the mean maximum round total?

```python
#Calculate the frequency of the number of ties within each game
df['Number_of_ties'].value_counts()
```

```python
#Calculate the mean maximum round total as a function of the number of ties
def mean_max_round_total(Number_of_ties):
    mean=df.loc[df['Number_of_ties'] ==Number_of_ties , 'Largest_total_won'].mean()
    return mean    
```

```python
print({'Number of ties = 0, the mean is':mean_max_round_total(0)})
print({'Number of ties = 1, the mean is':mean_max_round_total(1)})
print({'Number of ties = 2, the mean is':mean_max_round_total(2)})
print({'Number of ties = 3, the mean is':mean_max_round_total(3)})
print({'Number of ties = 4, the mean is':mean_max_round_total(4)})
```
