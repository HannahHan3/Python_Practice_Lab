a. (2 points) Import the Inc. 5000 data using basic file input techniques (i.e., either via open(filename) or the csv module). Create a dictionary (inc100) that contains the top 100 companies according to the rank (note that the rank is determine based on the growth of the companies revenues from the previous year), using the rank as the key. For each company (rank) in the dictionary, the value should be a dictionary containing the following (populated) fields:

'name': Company name (company)
'founded': Founding date (founded)
'industry': Industry (industry)
'metro': Metropolitan area of company location (metro)
'workers': Number of employees (workers)
Display (print) your full dictionary object and also demonstrate the ability to select specific companies from the dictionary according to their rank (at least two examples).

```python
#import data
import csv
filename = 'inc5000_2017.csv'
f = open(filename)
reader = csv.reader(f)
#create lists for each attributes
rows = []
name = []
rank = []
founded = []
industry = []
metro = []
workers = []

for line in reader:
    if line[4] in [str(i) for i in list(range(1,101))]:
        rows.append(line)
        name.append(line[-1])
        rank.append(line[4])
        founded.append(line[-2])
        industry.append(line[6])
        metro.append(line[3])
        workers.append(line[5])        
f.close()
```

```python
#build dictionary
dictionary={}
list_of_dict=[]

for i in range(0,101):
    dictionary['name']=name[i]
    dictionary['founded']=founded[i]
    dictionary['industry']=industry[i]
    dictionary['metro']=metro[i]
    dictionary['workers']=workers[i]
    list_of_dict.append(dictionary)
    dictionary={}
for i in range(0,101):
    dictionary[rank[i]]=list_of_dict[i]
```

Use the function to summarize the industry concentration for New York City, Atlanta, Chicago, and Washington (DC). What observations can you make about the industry concentration in each of these markets?

```python
#define function
def sum_industry(df,city='New York City'):
    data=df[df['city']==city]
    return data['industry'].value_counts()
```
